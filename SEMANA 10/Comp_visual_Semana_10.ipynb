Faça uma pesquisa sobre como é possível aplicar os operadores de gradiente Roberts, Prewitt e Sobel para detectar bordas em uma imagem, usando a linguagem Python.

Para cada operador, inclua o trecho de código que exemplifica a operação e um conjunto de quatro imagens (para cada operação) mostrando a imagem original, a imagem resultante da operação na horizontal, a imagem resultante da operação na vertical e a imagem representando a magnitude do gradiente.

Não se esqueça de incluir as referências na pesquisa.

```python
# LIBS
from skimage import data
from skimage.color import rgb2gray
import matplotlib.pyplot as plt
from skimage import filters
```

```python
#ROBERTS
original = data.astronaut()
img_grayscale = rgb2gray(original)

roberts_neg_d = filters.roberts_neg_diag(img_grayscale)
roberts_pos_d = filters.roberts_pos_diag(img_grayscale)
roberts_magnitude = filters.roberts(img_grayscale)

fig, axes = plt.subplots(1, 4, figsize=(12, 4))
ax = axes.ravel()

ax[0].imshow(img_grayscale, cmap=plt.cm.gray)
ax[0].set_title("Imagem Original")
ax[1].imshow(roberts_neg_d, cmap=plt.cm.gray)
ax[1].set_title("neg_diag")
ax[2].imshow(roberts_pos_d, cmap=plt.cm.gray)
ax[2].set_title("pos_diag")
ax[3].imshow(roberts_magnitude, cmap=plt.cm.gray)
ax[3].set_title("Sobel Magnitude do Gradiente")
fig.tight_layout()
plt.show()
```

```python
# SOBEL
original = data.astronaut()
img_grayscale = rgb2gray(original)

sobel_horizontal = filters.sobel_h(img_grayscale)
sobel_vertical = filters.sobel_v(img_grayscale)
sobel_gradiente = filters.sobel(img_grayscale)

fig, axes = plt.subplots(1, 4, figsize=(12, 4))
ax = axes.ravel()

ax[0].imshow(img_grayscale, cmap=plt.cm.gray)
ax[0].set_title("Imagem Original")
ax[1].imshow(sobel_horizontal, cmap=plt.cm.gray)
ax[1].set_title("Sobel Horizontal")
ax[2].imshow(sobel_vertical, cmap=plt.cm.gray)
ax[2].set_title("Sobel Vertical")
ax[3].imshow(sobel_gradiente, cmap=plt.cm.gray)
ax[3].set_title("Sobel Magnitude do Gradiente")
fig.tight_layout()
plt.show()
```

```python
# Prewitt
prewitt_horizontal = filters.prewitt_h(img_grayscale)
prewitt_vertical = filters.prewitt_v(img_grayscale)
prewitt_gradiente = filters.prewitt(img_grayscale)

fig, axes = plt.subplots(1, 4, figsize=(12, 4))
ax = axes.ravel()

ax[0].imshow(img_grayscale, cmap=plt.cm.gray)
ax[0].set_title("Imagem Original")
ax[1].imshow(prewitt_horizontal, cmap=plt.cm.gray)
ax[1].set_title("Prewitt Horizontal")
ax[2].imshow(prewitt_vertical, cmap=plt.cm.gray)
ax[2].set_title("Prewitt Vertical")
ax[3].imshow(prewitt_gradiente, cmap=plt.cm.gray)
ax[3].set_title("Prewitt Magnitude do Gradiente")
fig.tight_layout()
plt.show()
```
